```python
#@title import libraries

import numpy as np

from tensorflow.keras.datasets import fashion_mnist

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.losses import kullback_leibler_divergence as kl_div
from tensorflow.keras.utils import to_categorical

from tensorflow.keras import backend as K
```

```python
#@title load data
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
```

```python
#@title normalize data
x_train = x_train / x_train.max()
x_test = x_test / x_test.max()

# check the range
x_train.max()
```

```python
#@title reshape data for MLP
x_train = x_train.reshape(-1, 784)
x_test = x_test.reshape(-1, 784)

# check the shape
x_train.shape
```

```python
#@title one hot encode target
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# check shape
y_train.shape
```

```python
#@title create simple model
model = Sequential()
model.add(Input(shape=(784,)))
model.add(Dense(500, activation='relu'))
model.add(Dense(500, activation='relu'))
model.add(Dense(256, activation='relu'))
model.add(Dense(10, activation='softmax'))

model.summary()
```

```python
# compile model with loss as KL divergence
model.compile(loss=kl_div,
              optimizer="adam",
              metrics=['accuracy'])
```

```python
#@title train model
model.fit(x_train, y_train,batch_size=256,epochs=10)
```

```python
#@title evaluate on validation data
score = model.evaluate(x_test, y_test)
print(f'Test loss: {score[0]} / Test accuracy: {score[1]}')
```
